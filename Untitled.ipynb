```python
def get_chain_labels(mol): 
    hv = oechem.OEHierView(mol)
    chains = []
    for chain in hv.GetChains():
        chains.append(chain.GetChainID())
    return chains


def get_num_ligands(mol): 
    count = 0
    hv = oechem.OEHierView(mol)
    for chain in hv.GetChains():
        for frag in chain.GetFragments():
            for res in frag.GetResidues():
                if res.GetResidueName()=='LIG': 
                    count += 1
    return count
```

```python
def get_chain_labels_pdb(pdb_path):
    pdb_lines = pdb_path.open('rt').readlines()
    return list(set([x[21] for x in pdb_lines if x.startswith('ATOM')]))
```
